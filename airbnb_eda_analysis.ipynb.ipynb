```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_excel("1730285881-Airbnb_Open_Data.xlsx")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
print("The Shape of Dataset\n")
df.shape
```

```python
# Null values in columns
df.isnull().sum()
```

```python
# Out of 102599 rows license has 102597 rows null so drop this column
df = df.drop('license', axis=1)
```

```python
# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns:")
print(cat_cols, "\n")

# Numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Numeric columns:")
print(num_cols, "\n")
```

```python
print("Duplicate rows in dataset")
duplicates = df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
df = df.drop_duplicates()
print(f"Dataset shape after removing duplicates: {df.shape}")
```

```python
print("Unique neighbourhood groups:")
print(df['neighbourhood group'].unique(), "\n")

print("Unique neighbourhoods:")
print(df['neighbourhood'].unique(), "\n")
```

```python
df['neighbourhood group'] = df['neighbourhood group'].replace({
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan'
})
```

```python
print("Unique neighbourhood groups:")
print(df['neighbourhood group'].unique(), "\n")
```

```python
df['price'].nunique()
```

```python
print("Total listings:", df.shape[0])
print("Number of neighbourhood groups:", df['neighbourhood group'].nunique())
print("Number of neighbourhoods:", df['neighbourhood'].nunique())
print("Average price:", df['price'].mean())
print("Maximum price:", df['price'].max())
print("Minimum price:", df['price'].min())
print("Average rating:", df['review rate number'].mean())
```

```python
print("Listings by neighbourhood group:")
print(df['neighbourhood group'].value_counts())
```

```python
print("Average price by neighbourhood group:")
print(df.groupby('neighbourhood group')['price'].mean())
```

```python
print("Most popular neighbourhood (by listings):")
print(df['neighbourhood'].value_counts().idxmax())
```

```python
print("Listings by room type:")
print(df['room type'].value_counts())
```

```python
print("Average price by room type:")
print(df.groupby('room type')['price'].mean())
```

```python
print("Average rating by room type:")
print(df.groupby('room type')['review rate number'].mean())
```

```python
print("Median price:", df['price'].median())
```

```python
print("Listings with most reviews:")
print(df[['NAME','number of reviews']].sort_values(by='number of reviews', ascending=False).head(10))

print("Average reviews per month:", df['reviews per month'].mean())
```

```python
print("Listings available 365 days a year:", df[df['availability 365'] == 365].shape[0])
print("Average availability by neighbourhood group:")
print(df.groupby('neighbourhood group')['availability 365'].mean())
```

```python
print("Verified hosts:", df['host_identity_verified'].value_counts())
print("Top hosts by number of listings:")
print(df[['host name','calculated host listings count']].sort_values(by='calculated host listings count', ascending=False).head(10))
```

```python
for col in num_cols:
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f"Histogram of {col}")
    
    plt.subplot(1,2,2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    
    plt.show()
```

```python
# Categorical columns plots

for col in cat_cols:
    plt.figure(figsize=(10,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index[:20])
    plt.title(f"Count plot of {col} (top 20 categories)")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='room type', y='price', data=df)
plt.title("Price Distribution by Room Type")
plt.show()
```

```python
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values()
avg_price.plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title("Average Price by Neighbourhood Group")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['reviews per month'].dropna(), bins=30, kde=True)
plt.title("Distribution of Reviews per Month")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['availability 365'], bins=30, kde=False)
plt.title("Availability Distribution (365 days)")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Price vs Number of Reviews
plt.figure(figsize=(8,5))
sns.scatterplot(x='number of reviews', y='price', data=df)
plt.title("Price vs Number of Reviews")
plt.show()

# Price vs Reviews per Month
plt.figure(figsize=(8,5))
sns.scatterplot(x='reviews per month', y='price', data=df)
plt.title("Price vs Reviews per Month")
plt.show()
```

```python
# Violin plot for Review Rate Number vs. Price
plt.subplot(1, 2, 1)
sns.violinplot(x='review rate number', y='price', data=df)
plt.title('Effect of Review Rate Number on Price')
plt.xlabel('Review Rate Number')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
# Violin plot for Host Identity Verified vs. Price
plt.subplot(1, 2, 2)
sns.violinplot(x='host_identity_verified', y='price', data=df)
plt.title('Effect of Host Identity Verified on Price')
plt.xlabel('Host Identity Verified')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
# Assuming construction_year is the column that denotes the year of construction
construction_year_counts = df['Construction year'].value_counts().sort_index()

# Line graph to show the years with maximum construction of rooms
plt.figure(figsize=(10, 6))
construction_year_counts.plot(kind='line', marker='o', color='green')
plt.title('Construction of Rooms Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Rooms Constructed')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for service price vs room price
plt.figure(figsize=(8, 6))
plt.scatter(df['service fee'], df['price'], alpha=0.5, color='blue')
plt.title('Relationship between Service Price and Room Price')
plt.xlabel('Service Price')
plt.ylabel('Room Price')
plt.grid(True)
plt.show()
```
